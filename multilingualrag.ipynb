```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sentence_transformers import SentenceTransformer
```

```python
apple_company = ["Apple stock has sky rocketed today", "Apple is the most valuable company", "Steve Jobs is the CEO of Apple"]
apple_fruit = ["An apple a day keeps doctor away", "Apple is very healthy", "Apple grows on trees."]
query1 = "How's the share market today?"
query2 = "What's for breakfast?"
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')
```

```python
embeddings_company = model.encode(apple_company)
embeddings_fruit = model.encode(apple_fruit)

diabetes_sentences = ["Diabetes is a chronic condition", "Hypertension can lead to heart disease", "Fever can be treated with paracetemol"]
embeddings_diabetes = model.encode(diabetes_sentences)
all_sentences = apple_company + apple_fruit + diabetes_sentences
embeddings_all = model.encode(all_sentences)
```

```python
def get_results(query):
  embeddings = model.encode(query)
  embeddings = embeddings.reshape(1, -1)
  cosine_sim = cosine_similarity(embeddings, embeddings_all).flatten()
  # Combine sentences with their similarity scores
  sentence_score_pairs = list(zip(all_sentences, cosine_sim))

  # Sort the pairs by similarity score in descending order
  sentence_score_pairs_sorted = sorted(sentence_score_pairs, key=lambda x: x[1], reverse=True)

  # Output results
  print(f"Query : {query}\n")
  print("Sentences ranked by similarity:\n")
  for sentence, score in sentence_score_pairs_sorted:
      print(f"'{sentence}' with a similarity score of {score:.4f}")


```

```python
get_results("What's for breakfast?")
```

```python
get_results("ما هو لتناول الافطار؟")
```
